## Data preprocessing

### Remove unnecessary content (tag, link,emoticon,etc)

```python
def clean_contents(content, exceptions):
    # Create patterns for mentions, URLs, emoticons, hashtags, and other characters
    mention_pattern = re.compile(r'@\w+')
    url_pattern = re.compile(r'https?://\S+')
    emoticon_pattern = re.compile(r'\S*?(\:\)|\:\(|\:D|\;\)|\<3|\:\*\))')
    hashtag_pattern = re.compile(r'#\w+')
    other_pattern = re.compile(r'[^\s\w@]|_|\b\d+\b')

    # Combine all patterns
    combined_pattern = re.compile(f'({mention_pattern.pattern}|{url_pattern.pattern}|{emoticon_pattern.pattern}|{hashtag_pattern.pattern}|{other_pattern.pattern})')

    # Use re.sub to replace matching patterns with an empty string
    cleaned_text = combined_pattern.sub(lambda match: match.group(1) if match.group(1) and match.group(1) in exceptions else '', content)

    # Remove unnecessary spaces
    cleaned_text = re.sub(r'\s+', ' ', cleaned_text).strip()

    return cleaned_text


```

```python
exception_list = ['gibran', 'ganjar', 'prabowo', 'anies','PakbowoGibran', 'Gangar']
cleaned_contents_np = np.array([])
for content in tweets_df['content']:
    cleaned_content = clean_contents(content, exception_list)
    cleaned_contents_np = np.append(cleaned_contents_np, cleaned_content)
```

### Data Labelazitaion

```python
from google.cloud import language_v1

def analyze_sentiment(text_content):
    client = language_v1.LanguageServiceClient()

    # Available types: PLAIN_TEXT
    type_ = language_v1.Document.Type.PLAIN_TEXT

    # Optional. If not specified, the language is automatically detected.
    language = "id"
    document = {"content": text_content, "type_": type_, "language": language}

    # Available values: NONE, UTF8, UTF16, UTF32
    encoding_type = language_v1.EncodingType.UTF32


    response = client.analyze_sentiment(request={'document': document, 'encoding_type': encoding_type})

    # Get the overall sentiment of the input document
    sentiment = response.document_sentiment

    return sentiment.score, sentiment.magnitude
```

```python
scores= []
magnitudes = []
sentiments = []
for content in stemmed_contents_np:
    score, magnitude = analyze_sentiment(content)
    scores.append(score)
    magnitudes.append(magnitude)
    sentiment = "Positive" if score > 0 else "Neutral" if score == 0 else "Negative"
    sentiments.append(sentiment)

tweets_df['score'] = scores
tweets_df['magnitude'] = magnitudes
tweets_df['sentiment'] = sentiments
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf_vectorizer = TfidfVectorizer()
# Create the TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer()

# Fit and transform the documents
tfidf_matrix = tfidf_vectorizer.fit_transform(stemmed_contents_np)

# Get the feature names (words)
feature_names = tfidf_vectorizer.get_feature_names_out()

# Create a DataFrame to display the TF-IDF matrix
df_tfidf = pd.DataFrame(data=tfidf_matrix.toarray(), columns=feature_names)
# print(df_tfidf)
np_tfidf = df_tfidf.to_numpy()
print(np_tfidf.shape)
```
